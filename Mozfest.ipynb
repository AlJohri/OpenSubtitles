Load Movie Data

```python
from load import movies
movies = np.array(movies)
print ("loaded data")
```

```python
N_CLUSTERS = 3
k_means = KMeans(n_clusters=N_CLUSTERS, init='k-means++', max_iter=100, n_init=1, verbose=True)
vectorizer = TfidfVectorizer(max_df=0.5, min_df=2, stop_words='english')
```

```python
text = [movie['script'] for movie in movies]
X = vectorizer.fit_transform(text)
```

```python
km = k_means.fit(X)
```

```python
k_means_labels = k_means.labels_
k_means_cluster_centers = k_means.cluster_centers_
k_means_labels_unique = np.unique(k_means_labels)
```

```python
cluster0 = [movie['script'] for movie in movies[k_means_labels == 0]]
vectorizer2 = TfidfVectorizer(max_df=0.5, min_df=2, stop_words='english')
z = vectorizer2.fit_transform(cluster0)
wordz = vectorizer2.get_feature_names()
wordz_tfidf = [(wordz[i], z[:,i].sum()) for i in range(z.shape[1])]
wordz_tfidf = sorted(wordz_tfidf, key=lambda x: x[1], reverse=True )
wordz_tfidf
```

```python
order_centroids = k_means_cluster_centers.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
for ind in order_centroids[0, :100]:
    print(' %s' % terms[ind], end='')
```
